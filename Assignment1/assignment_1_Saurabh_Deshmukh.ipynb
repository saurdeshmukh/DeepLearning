```python
def costFunction(X,W,B,Y):
    m=len(Y)
    sumE=0
    for i in range(0,m):
        sumE=sumE+(X[i]*W+B-Y[i])
    return sumE/m
```

```python
def RootMeanSquareError(Y,Y_):
    SUM=0
    for i in range(0,len(Y)):
        SUM+=(Y[i]-Y_[i])**2
    return np.sqrt((SUM/len(Y)))
        
```

```python
X11tf=np.vstack((X88,X77,X66,X55,X44,X33,X22,X11)).T
```

```python
def append_bias_reshape(X_,Y_):
    n_training_samples = X_.shape[0]
    n_dim = X_.shape[1]
    f = np.reshape(np.c_[np.ones(n_training_samples),X_],[n_training_samples,n_dim + 1])
    l = np.reshape(Y_,[n_training_samples,1])
    return f, l
```

```python
Xtf,Ytf=append_bias_reshape(X11tf,Y11)
```

```python
Xtf.shape
```

```python
Ytf.shape
```

```python
lr=0.01
```

```python
train_epochs=10000
```

```python
cost_history = np.empty(shape=[1],dtype=float)
n_dim = Xtf.shape[1]
X_tf = tf.placeholder(tf.float32,[None,n_dim])
Y_tf = tf.placeholder(tf.float32,[None,1])
W_tf = tf.Variable(tf.ones([n_dim,1]))

init = tf.global_variables_initializer()
```

```python
y_ = tf.matmul(X_tf, W_tf)
cost = tf.reduce_mean(tf.square(y_ - Y_tf))
training_step = tf.train.GradientDescentOptimizer(lr).minimize(cost)
```

```python
sess = tf.Session()
sess.run(init)

for epoch in range(train_epochs):
    sess.run(training_step,feed_dict={X_tf:Xtf,Y_tf:Ytf})
    cost_history = np.append(cost_history,sess.run(cost,feed_dict={X_tf:Xtf,Y_tf:Ytf}))
```

```python
plt.plot(range(len(cost_history)),cost_history)
plt.axis([0,train_epochs,0,np.max(cost_history)])
plt.show()
```
